```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
def heart_d(r):
  if r==0:
    return 'Absence'
  else:
    return 'Presence'

df['Heart_Disease']=df['target'].apply(heart_d)
df.head(10)
```

```python
hd=df.groupby('Heart_Disease')['target'].count()
hd
```

```python
plt.figure(figsize=(10,7))
plt.pie(hd,labels=['Absence','Presence'],autopct='%0.0f%%')
plt.title('Heart Disease Population %',fontsize=20)
plt.show()

```

```python
fig, ax = plt.subplots(1, 2)
age = df['age'].values
sex = df['sex'].values

sns.histplot(age, ax=ax[0], color='blue', kde=True)
ax[0].set_title('Age Distribution')
ax[0].set_xlim([min(age), max(age)])

sns.histplot(sex, ax=ax[1], color='orange', kde=True)
ax[1].set_title('Sex Distribution')
ax[1].set_xlim([min(sex), max(sex)])

plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='age',data=df)
plt.title("Population Age",fontsize=17)
plt.xlabel("Age",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
min_age=df['age'].min()
max_age=df['age'].max()
mean_age=df['age'].mean()
print("Minimum Age = ",min_age)
print("Maximum Age = ",max_age)
print("Meanimum Age = ",mean_age)
```

```python
Young_Ages=df[(df['age']>=29) & (df['age']<40)]
Middle_Ages=df[(df['age']>=40) & (df['age']<55)]
Elderly_Ages=df[(df['age']>=55)]
print("Young Ages=",len(Young_Ages))
print("Middle Ages=",len(Middle_Ages))
print("Elderly Ages=",len(Elderly_Ages))
```

```python
def gender(r):
  if r==1:
    return "Male"
  else:
    return "Female"

df['sex1']=df['sex'].apply(gender)
df.head()
```

```python
def age_range(r):
  if r>=29 and r<40:
    return "Young Age"
  elif r>=40 and r<55:
    return "Middle Age"
  else:
    return "Elder Age"

df['Age_Range']=df['age'].apply(age_range)
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.stripplot(x='Age_Range', y='age', hue='sex1', data=df, order=['Young Age', 'Middle Age', 'Elder Age'], palette='Oranges_r', jitter=True, dodge=True)
plt.title('Gender Based Age Category', fontsize=17)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,5))
hue_order=['Young Age','Middle Age','Elder Age']
sns.countplot(x='Heart_Disease',hue='Age_Range',data=df,order=['Presence','Absence'],hue_order=hue_order,palette='Pastel1')
plt.title('Heart Disease Based on Age Category',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x=df['Heart_Disease'],hue='sex1',data=df,palette='BuGn_r')
plt.title('Heart Disease Based on Gender',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Male','Female'])
plt.show()
```

```python
sns.countplot(x=df['Heart_Disease'],hue='cp',data=df,order=['Presence','Absence'])
plt.title('Chest Pain Explained',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```
